```python
import os
from PIL import Image

def convert_gif_to_jpg(gif_path, jpg_folder):
    try:
        gif_image = Image.open(gif_path)
        while True:
            try:
                current_frame = gif_image.tell()
                gif_image.seek(current_frame)
                jpg_path = os.path.join(jpg_folder, f"{os.path.splitext(os.path.basename(gif_path))[0]}_{current_frame}.jpg")
                jpg_image = gif_image.convert("RGB")
                jpg_image.save(jpg_path)
                print(f"Successfully converted frame {current_frame} of '{gif_path}' to '{jpg_path}'.")
                gif_image.seek(current_frame + 1)
            except EOFError:
                break

    except Exception as e:
        print(f"Error converting '{gif_path}' to JPG: {e}")

def batch_convert_gif_to_jpg(gif_folder, jpg_folder):
    if not os.path.exists(jpg_folder):
        os.makedirs(jpg_folder)

    gif_files = [file for file in os.listdir(gif_folder) if file.endswith(".gif")]

    for gif_file in gif_files:
        gif_path = os.path.join(gif_folder, gif_file)
        convert_gif_to_jpg(gif_path, jpg_folder)

gif_folder_path = "D04-Hazy weather"
jpg_folder_path = "D04_images"
batch_convert_gif_to_jpg(gif_folder_path, jpg_folder_path)
convert_gif_to_jpg(gif_folder_path, jpg_folder_path)
```
